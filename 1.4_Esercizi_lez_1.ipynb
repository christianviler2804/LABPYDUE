### **1. Trasformare cicli in list comprehension**

Fai tre esempi di cicli producono una lista e poi riscrivi lo stesso codice usando la list comprehension

```python
a = []
for i in range(1, 4):
    a.append(i)
b = []
for i in range(1, 10, 2):
    b.append(i**i)
c = []
for i in "stringa":
    c.append(i)

a_comp = [i for i in range(1, 4)]
b_comp = [i**i for i in range(1, 10, 2)]
c_comp = [i for i in "stringa"]

print(a, a_comp)
print(b, b_comp)
print(c, c_comp)
```

1. **Crea un vettore** contenente tutti i numeri primi compresi tra **0 e 10**

3. Quale pensi sia il tipo di dato (`dtype`) del vettore?  Prova a rispondere senza eseguire il codice e Verifica la tua risposta accedendo all'attributo `.dtype` del vettore

4. Scrivi l'array usando una list comprehension che controlla che i numeri siano primi. 

```python
import numpy as np
def verifica_primo (n: int):
    r = np.array([])
    for i in range(2, n):
        cont = 0
        for c in range(2, (i//2)+1):
            if (i%c == 0):
                cont+=1
        if (cont == 0):
            r = np.append(r, i)
    r = r.astype(int)
    return r

b = verifica_primo(10)
a = np.array([2, 3, 5, 7])
print(a)
print(b)

print(len(a))
print(a.size)

print(a.dtype)

def primo (i:int):
    if (i == 0 or i == 1):
        return False
    cont = 0
    for c in range(2, (i//2)+1):
            if (i%c == 0):
                cont+=1
    if (cont == 0):
        return True
    else:
        return False
    
v = np.array([i for i in range(11) if primo(i)])
print(v)
```
